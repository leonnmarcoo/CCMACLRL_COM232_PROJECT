# **Data Visualization**

### **Import**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### **Load Dataset (Original Data)**

```python
df = pd.read_csv('../cleaned_data.csv')
```

```python
df.head()
```

```python
df.info()
```

### **Data Visualization**

```python
df['Addicted_Score'].value_counts().plot(kind='bar')
plt.title('Distribution of Addicted Scores')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].scatter(df["Age"], df["Addicted_Score"])
ax[0, 1].scatter(df["Gender"], df["Addicted_Score"])
ax[1, 0].scatter(df["Academic_Level"], df["Addicted_Score"])
ax[1, 1].scatter(df["Country"], df["Addicted_Score"])

ax[0, 0].set_title(["Age and Addicted_Score"])
ax[0, 1].set_title(["Gender and Addicted_Score"])
ax[1, 0].set_title(["Academic_Level and Addicted_Score"])
ax[1, 1].set_title(["Country and Addicted_Score"])
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].scatter(df["Avg_Daily_Usage_Hours"], df["Addicted_Score"])
ax[0, 1].scatter(df["Most_Used_Platform"], df["Addicted_Score"])
ax[1, 0].scatter(df["Affects_Academic_Performance"], df["Addicted_Score"])
ax[1, 1].scatter(df["Sleep_Hours_Per_Night"], df["Addicted_Score"])

ax[0, 0].set_title(["Avg_Daily_Usage_Hours and Addicted_Score"])
ax[0, 1].set_title(["Most_Used_Platform and Addicted_Score"])
ax[1, 0].set_title(["Affects_Academic_Performance and Addicted_Score"])
ax[1, 1].set_title(["Sleep_Hours_Per_Night and Addicted_Score"])
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].scatter(df["Mental_Health_Score"], df["Addicted_Score"])
ax[0, 1].scatter(df["Relationship_Status"], df["Addicted_Score"])
ax[1, 0].scatter(df["Conflicts_Over_Social_Media"], df["Addicted_Score"])
ax[1, 1].scatter(df["Addicted_Score"], df["Addicted_Score"])

ax[0, 0].set_title(["Mental_Health_Score and Addicted_Score"])
ax[0, 1].set_title(["Relationship_Status and Addicted_Score"])
ax[1, 0].set_title(["Conflicts_Over_Social_Media and Addicted_Score"])
ax[1, 1].set_title(["Addicted_Score and Addicted_Score"])
```

```python
correlation = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, vmin=-1, vmax=1, center=0, fmt=".3f")
plt.show
```
